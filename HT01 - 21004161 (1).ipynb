### Problema 1 

Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos:
   - Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul.
   - Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde.
   - Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo.
   - Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
   - Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul.
   - Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.

```python
#Definición de la función RGBFIlter, con parámetros imagen y entero

def RGBFilter(imagen,color):
    img=imagen
    alto=img.shape[0]
    ancho=img.shape[1]
    color1 = np.zeros((alto,ancho,3))  #azul
    color2 = np.zeros((alto,ancho,3))  #verde
    color3 = np.zeros((alto,ancho,3))  #rojo
    color10 = np.zeros((alto,ancho,3)) #rojo y verde
    color20 = np.zeros((alto,ancho,3)) #verde y azul
    color30 = np.zeros((alto,ancho,3)) #azul y rojo
    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]
            red=pixel[0]
            green=pixel[1]
            blue=pixel[2]
            if color==1:
                color1[i,j]=[0,0,blue]
            elif color==2:
                color2[i,j]=[0,green,0]
            elif color==3:
                color3[i,j]=[red,0,0]
            elif color==10:
                color10[i,j]=[red,green,0]
            elif color==20:
                color20[i,j]=[0,green,blue]
            elif color==30:
                color30[i,j]=[red,0,blue]
    if color==1:
        return color1
    elif color==2:
        return color2
    elif color==3:
        return color3
    elif color==10:
        return color10
    elif color==20:
        return color20
    elif color==30:
        return color30
    end
```

### Problema 3
Cree una función que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su función debe crear 3 imágenes como salida. Para entregar este ejercicio debe incluir una de las imágenes que haya utilizado como prueba y los resultados de la misma, no puede utilizar la imagen del Problema #2.

```python
def EscalaGrises(img):
    alto=img.shape[0]
    ancho=img.shape[1]   
    grisR = np.zeros((alto,ancho,3))
    grisG = np.zeros((alto,ancho,3))
    grisB = np.zeros((alto,ancho,3))
    for i in range(0,alto):
        for j in range(0,ancho):
            pixel=img[i,j]
            red=pixel[0]
            green=pixel[1]
            blue=pixel[2]
            grisR[i,j]=[red,red,red]
            grisG[i,j]=[green,green,green]
            grisB[i,j]=[blue,blue,blue]
    return grisR,grisG,grisB
```

# <h1><center><font color=blue>Modelo HSV</font>  

El modelo de color HSV es una transformación no lineal del modelo RGB en coordenadas cilíndricas de manera que cada color viene definido por las siguientes dimensiones:

   - -H- **Tinte o matiz**: Ángulo que representa el matiz, normalmente definido entre 0o y 360o. Las amplitudes para cada tonalidad de color abarcan 120o. Por ejemplo los rojizos (255,0,0) se disponen a ambos lados de 0o, mientras los verdes (0,255,0) lo hacen alrededor de 120o y por último los tonos azules (0,0,255) se distribuyen alrededor de 240o.
   - -S- **Saturación**: Nivel saturación del color, dado entre 0 y 1, 0 representa sin saturación alguna (blanco), hasta 1 que sería el matiz en toda su intensidad. Es común también darlo en percentiles 0%-100%.
   - -V- **Brillo**: Nivel del brillo entre 0 y 1. 0 es negro; 1, blanco. Al igual que la saturación puede darse en porcientos entre 0% y 100%. De esta forma el 50% indica el nivel medio o normal del brillo del color.
   
Viene del inglés (Hue, Saturation, Value) = (Matiz, Saturación, Valor)
