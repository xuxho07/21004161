```python
import pandas as pd
import os
from pathlib import Path

os.getcwd()
!pwd
%cd /Users/xuxho/Desktop/
!pwd

files = [file for file in os.listdir("./Sales_Data")]
print(files)

completeData = pd.DataFrame()
for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    completeData=pd.concat([completeData, df])
completeData.head(7)
#[  ::-1   ]#
```

```python
completeData[completeData['Order Date'].isna()]
```

```python
completeData['Order Date'].str[0:2].unique()
completeData.shape
```

```python
completeData = completeData[completeData['Order Date'].str[0:2] != 'Or']
completeData.shape
```

```python
completeData.dropna(how="all", inplace=True)
completeData.shape
```

```python
completeData["Month"]= completeData['Order Date'].str[0:2]
completeData.shape
#completeData["Month"]=pd.to_numeric(df["Month"], downcast="integer")
#completeData.head()
```

```python
completeData["Month"]=pd.to_numeric(completeData["Month"], downcast="integer")
completeData.head()
```

```python
completeData["Year"] = pd.DatetimeIndex(completeData['Order Date']).year
completeData.head()
```

```python
completeData=completeData[completeData["Year"]==2019]
completeData.shape
```

```python
completeData["Quantity Ordered"]=pd.to_numeric(completeData["Quantity Ordered"], downcast="integer")
completeData["Price Each"]=pd.to_numeric(completeData["Price Each"], downcast="float")
completeData.head()
```

```python
completeData["Sales"]=completeData["Quantity Ordered"]*completeData["Price Each"]
completeData.head()
```

```python
completeData.groupby("Month").sum()
```

```python
def getCity(x):
    return x.split(", ")[1].strip()

def getState(x):
    return x.split(", ")[2].strip(' ')[1]

completeData["Ciudad"]= completeData['Purchase Address'].apply(lambda x: getCity(x)+" ("+getState(x)+")")
completeData.head()
```

```python
citiesData=completeData.groupby("Ciudad").sum()
citiesData.head()
```
